```
# Consultas INNER JOIN en python panda


inner_join = dfTransaccion.merge(dfUsuario, on='IdUsuario', how='inner')
# Filtrar por IdUsuario = 5
Consulta4 = inner_join[inner_join['IdUsuario'] == 2]
print(Consulta4.to_string(index=False))
print("\n")


# Filtrar por IdUsuario = 3
Consulta5 = inner_join[inner_join['IdUsuario'] == 3]
print(Consulta5.to_string(index=False))
print("\n")

# Filtrar por IdUsuario = 1
Consulta6 = inner_join[inner_join['IdUsuario'] == 1]
print(Consulta6.to_string(index=False))
print("\n")
```

```
# Consultas UNION en python panda

# traer los id de los usuarios y nombres que han realizado transacciones
Consulta7 = pd.merge(
    dfUsuario[['IdUsuario', 'Nombre']],  # columnas de usuario
    dfTransaccion[['IdUsuario']],        # columnas de transaccion
    on='IdUsuario',                      # columna clave para unir
    how='inner'                          # INNER JOIN
)
print(Consulta7)
print("\n")

#traer los id de los usuarios con los id de los cupones usados y sus valores

Consulta8 = pd.merge(
    dfTransaccion[['IdUsuario', 'IdCupon']],  # tomamos solo estas columnas
    dfCupon[['IdCupon', 'Valor']],           # unimos con valor del cupón
    on='IdCupon',                             # clave de unión
    how='inner'                               # INNER JOIN
)

print(Consulta8)
print("\n")

# traer IdUsuario, IdResiduo y Nombre del usuario y fecha en la que se llevo el residuo
Consulta9 = pd.merge(
    dfTransaccion[['IdUsuario', 'IdResiduo', 'Fecha']],  # columnas de Transaccion
    dfUsuario[['IdUsuario', 'Nombre']],                  # columnas de Usuario
    on='IdUsuario',                                     # clave de unión
    how='inner'                                        # INNER JOIN
)

print(Consulta9)
```
